# 5. 普遍性の証明

```python
"""色々準備"""
from qiskit import QuantumCircuit
from math import pi
```

```python
"""
1量子ビット Rx(θ) と2つの制御NOTからユニタリー行列 e^{i θ/2 X⊗X⊗X} を作成する方法を知っています。
"""
qc = QuantumCircuit(3)

theta = pi/2

qc.cx(0,2)
qc.cx(0,1)
qc.rx(theta,0)
qc.cx(0,1)
qc.cx(0,2)

qc.draw()
```

```python
"""
いくつかのアダマールがあれば、 e{i θ/2 Z⊗Z⊗Z} にも同じことができます。
"""
qc = QuantumCircuit(3)

theta = pi/2

qc.h(0)
qc.h(1)
qc.h(2)
qc.cx(0,2)
qc.cx(0,1)
qc.rx(theta,0)
qc.cx(0,1)
qc.cx(0,2)
qc.h(2)
qc.h(1)
qc.h(0)

qc.draw()
```
